# Solution ðŸ’™ 21. (73) Set Matrix Zeroes

## ðŸ“ Quick explanation

- Set rows-length and column-length (`m, n = len(matrix)m len(matrix[0])`)

- We establish `setRows, setCols = [0]*m, [0]*n` which will determine which rows and columns need to be set to 0

- Then iterate over the whole matrix, if a 0 is found, `setRows[r], setCols[c] = 1, 1`.

- With setRows and setCols, we define 2 for loops that will replace the values to 0 for a given row or column:

```python
        # set rows to 0
        for r in range(m):
            if setRows[r] == 1:
                for c in range(n):
                    matrix[r][c]=0

        # set columns to 0
        for c in range(n):
            if setCols[c] == 1:
                for r in range(m):
                    matrix[r][c]=0
```


## ðŸ’™ Code
```python
class Solution(object):
    def setZeroes(self, matrix):
        m , n = len(matrix) , len(matrix[0])
        setRows , setCols = [0] * m ,[0]* n

        for r in range(m):
            for c in range(n):
                if matrix[r][c]==0:
                    setRows[r] , setCols[c] = 1 , 1
        for r in range(m):
            if setRows[r] == 1:
                for c in range(n):
                    matrix[r][c]=0
        
        for c in range(n):
            if setCols[c] == 1:
                for r in range(m):
                    matrix[r][c]=0

        return matrix
```

## ðŸ’™ Another solution (neetcode)

```python
class Solution(object):
    def setZeroes(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: None Do not return anything, modify matrix in-place instead.
        """
        # O(1)
        rows, cols = len(matrix), len(matrix[0])
        rowZero = False

        # determine which rows/cols need to be zero
        for r in range(rows):
            for c in range(cols):
                if matrix[r][c] == 0:
                    matrix[0][c] = 0
                    
                    if r > 0:
                        matrix[r][0] = 0
                    else:
                        rowZero = True
        
        for r in range(1, rows):
            for c in range(1, cols):
                if matrix[0][c] == 0 or matrix[r][0] == 0:
                    matrix[r][c] = 0
        if matrix[0][0] == 0:
            for r in range(rows):
                matrix[r][0] = 0
        
        if rowZero:
            for c in range(cols):
                matrix[0][c] = 0
        
```
