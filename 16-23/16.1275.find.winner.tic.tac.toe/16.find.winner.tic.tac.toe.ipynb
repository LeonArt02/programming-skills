### 1275. Find Winner on a Tic Tac Toe Game

Tic-tac-toe is played by two players `A` and `B` on a `3 x 3` grid. The rules of Tic-Tac-Toe are:

- Players take turns placing characters into empty squares `' '`.
- The first player `A` always palcers `'X'`, while the second player `B'` always places `'O'` characters.
- `'X'` and `'O'` characters are always placed into empty squares, never on filled ones.
- The game ends when there are **three** of the same (non-empty) character filling any row, column, or diagonal.
- The game also ends if all squares are non-empty.
- No more moves can be played if the game is over.

**Example 1:**



- Input: moves = [[0,0],[2,0],[1,1],[2,1],[2,2]]
- Output: "A"
- Explanation: A wins, they always play first.


**Example 2:**



- Input: moves = [[0,0],[1,1],[0,1],[0,2],[1,0],[2,0]]
- Output: "B"
- Explanation: B wins.


**Example 3:**



- Input: moves = [[0,0],[1,1],[2,0],[1,0],[1,2],[2,1],[0,1],[0,2],[2,2]]
- Output: "Draw"
- Explanation: The game ends in a draw since there are no moves to make.



Given a 2D integer array **`moves`** where **`moves[i] = [rowi, coli]`** indicates that the **`i'th`** move will be played on `grid[rowi][coli]`. return the *winner* of the game if it exists (*A* or *B*). In case the game ends in a draw return `"Draw"`. If there are still movements to play return `"Pending"`.

You can asume that `moves` is valid (i.e., it follows the rules of Tic-Tac-Toe), the grid is initially empty, and A will play first.

```python
class Solution(object):
    def tictactoe(self, moves):
        """
        :type moves: List[List[int]]
        :rtype: str
        """
        a = [0,0,0,0,0,0,0,0]
        b = [0,0,0,0,0,0,0,0]
        
        for i in range(len(moves)):
            row, col = moves[i]
            if i%2==0:
                player = a
            else:
                player = b
            #row
            player[row] += 1
            # column
            player[col + 3] += 1
            # 1 diag
            if row == col:
                player[6] += 1
            # 2 diag
            if row == 2-col:
                player[7] += 1 
        if 3 in a:
            return 'A'
        if 3 in b:
            return 'B'
        
        if len(moves) == 9:
            return 'Draw'
        else:
            return 'Pending'
```

```python
"""
Input: moves = [[0,0],[2,0],[1,1],[2,1],[2,2]]
Output: "A"
Explanation: A wins, they always play first.
"""
sol = Solution()
print(sol.tictactoe(moves = [[0,0],[2,0],[1,1],[2,1],[2,2]]))
# Example 1 expected output: 'A' | Solved ✅

print(sol.tictactoe(moves = [[0,0],[1,1],[0,1],[0,2],[1,0],[2,0]]))
# Example 2 expected output: 'B' | Solved ✅

print(sol.tictactoe(moves = [[0,0],[1,1],[2,0],[1,0],[1,2],[2,1],[0,1],[0,2],[2,2]]))
# Example 3 expected output: 'Draw' | Solved ✅
```
